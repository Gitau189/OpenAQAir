Plotting the Box PLOT

```python
def plot_boxplot(df, x, y, legend=None):
  """
  Plots a box plot of the specified data, optionally displaying actual parameter names on the x-axis.

  Args:
      df (pandas.DataFrame): The DataFrame containing the data.
      x (str): The column name for the x-axis (grouping variable).
      y (str): The column name for the quantitative variable to plot.
      legend (str, optional): The column name containing parameter names (for labeling). Defaults to None.
  """

  plt.figure(figsize=(8, 6))

  plt.boxplot(
      [df[df[x] == group][y] for group in df[x].unique()]
  )
  
  if legend is not None:
      x_labels = df[legend].unique()
      x_labels = [label.upper() for label in x_labels]
      plt.xticks(range(1, len(x_labels) + 1), x_labels)
  else:
      plt.xlabel(x.capitalize())

  plt.ylabel(y.capitalize())
  plt.title(f'Box Plot of {y.capitalize()} by {x.capitalize()}')
  plt.tight_layout()
  plt.show()
```

```python
x = 'parameter'
y = 'value'
legend = 'parameter'
```

```python
plot_boxplot(tokyo, x, y, legend)
```

## Ploting the line graph of the parameters against the dates


```python
tokyo['datetimeUtc']=pd.to_datetime(tokyo['datetimeUtc'])
```

```python
def plot_line_graph(df, x_axis, y_axis, legend=None):
    """
    Plots a line graph of every value disaggregated by the 'parameter' column.

    Args:
        df (pandas.DataFrame): The DataFrame containing the data.
        x (str): The column name for the x-axis (grouping variable).
        y (str): The column name for the quantitative variable to plot.
        legend (str, optional): The column name containing parameter names (for labeling). Defaults to None.
    """

    plt.figure(figsize=(10, 6))
    for parameter in df[legend].unique():
        df_filtered = df[df[legend] == parameter]
        plt.plot(df_filtered[x_axis], df_filtered[y_axis], label=parameter)

   
    plt.ylabel(y_axis.capitalize())
    plt.title(f'Line Graph Disaggregated by {legend.capitalize()}')
    if legend:
        plt.legend()
    plt.grid(True)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
x_axis = 'datetimeUtc'
y_axis = 'value'
legend = 'parameter'
```

```python
plot_line_graph(tokyo, x_axis, y_axis, legend)
```

## Plotting each parameter against the dates

```python
tokyo['date']= tokyo['datetimeUtc'].dt.to_period('D')

tokyo.drop(['datetimeUtc'], axis=1, inplace=True)
tokyo
```

```python

df_no = tokyo[tokyo['parameter'] == 'no'].copy()  # Filter the data for 'pm25' measurements and create a copy

# Calculate the daily average of 'no' values
df_no.loc[:, 'daily_avg_no'] = df_no.groupby('date')['value'].transform('mean')

#calculate the daily average of 'no2'
df_no2 = tokyo[tokyo['parameter'] == 'no2'].copy()
df_no2.loc[:, 'daily_avg_no2'] = df_no2.groupby('date')['value'].transform('mean')
df_no2
#calculate the daily average of 'nox'
df_nox= tokyo[tokyo['parameter'] == 'nox'].copy()
df_nox.loc[:, 'avg_nox'] = tokyo.groupby('date')['value'].transform('mean')
df_nox

#calculate the daily average of 'so2'
df_so2= tokyo[tokyo['parameter'] == 'so2'].copy()
df_so2.loc[:, 'avg_so2'] = tokyo.groupby('date')['value'].transform('mean')
df_so2


```

```python


df_no.to_csv('no.csv', index=False)
df_no2.to_csv('no2.csv', index=False)
df_nox.to_csv('nox.csv', index=False)
df_so2.to_csv('so2.csv', index=False)
```
